```python
ratings = pd.read_csv('./data/ratings_small.csv')
movies = pd.read_csv('./data/movie_dataset.csv')
```

```python
movies.columns
```

```python
movies.rename(columns = {'id': 'movieId'}, inplace = True)
```

```python
ratings_movies = pd.merge(ratings,movies, on='movieId')
```

```python
ratings_movies.head(1)
```

```python
ratings_movies.shape
```

```python
data = ratings_movies.pivot_table('rating', index='userId', columns='title').fillna(0)
```

```python
data.head()
```

```python
data.shape
```

- 위와 같이 사용자별 영화 title에 따라 평점을 매긴 정보를 볼 수 있게 된다.
- 문제점 : 아이템 기반 협업 필터링이므로 row가 user 기반이면 안됨
- item이 row가 되어야함. -> 유사도 측정 

```python
data = data.transpose()
data.head(2)
```

```python
data.shape
```

```python
movie_sim = cosine_similarity(data,data)
print(movie_sim.shape)
```

```python
movie_sim_df = pd.DataFrame(data=movie_sim, index= data.index, columns= data.index)
```

```python
movie_sim_df
```

특정 영화와 비교했을 때 그 영화와 유사한 영화들을 추천 해주면 된다

```python
movie_sim_df["10 Things I Hate About You"].sort_values(ascending=False)[1:10]
```
